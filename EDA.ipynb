## Features

### Replacing none values

```python
data['Feature_1'].fillna(value='Power Steering',inplace=True)
data['Feature_2'].fillna(value='Power Windows Front',inplace=True)
data['Feature_3'].fillna(value='Air Conditioner',inplace=True)
data['Feature_4'].fillna(value='Heater',inplace=True)
data['Feature_5'].fillna(value='Adjustable Head Lights',inplace=True)
data['Feature_6'].fillna(value='Fog Lights Front',inplace=True)
data['Feature_7'].fillna(value='Anti Lock Braking System',inplace=True)
data['Feature_8'].fillna(value='Centeral Locking',inplace=True)
data['Feature_9'].fillna(value='Cd Player',inplace=True)
```

```python
data.isnull().sum()
```

## body type

```python
bodytype = data[data['bt'].isna()]
bodytype
```

```python
data['bt'].unique()
```

```python
data['bt'] = data.apply(lambda row: 'MUV' if pd.isnull(row['bt']) and row['oem'] == 'Toyota' else row['bt'], axis=1)
data['bt'] = data.apply(lambda row: 'Minivans' if pd.isnull(row['bt']) and row['oem'] == 'Maruti' else row['bt'], axis=1)
```

## Engine CC

```python
cc = data[data['Engine CC'].isna()]
cc
```

```python
data['Engine CC'].fillna(value=1300,inplace=True)
```

## Seater

```python
seats = data[data['seater'].isna()]
seats['bt'].unique()
```

```python
for i in data.iterrows():
    if i[1]['bt'] == 'Hatchback':
        data['seater'] = data['seater'].apply(lambda x: '3' if pd.isnull(x) else x)
    elif i[1]['bt'] == 'SUV':
        data['seater'] = data['seater'].apply(lambda x: '5' if pd.isnull(x) else x)
    elif i[1]['bt'] == 'MUV':
        data['seater'] = data['seater'].apply(lambda x: '7' if pd.isnull(x) else x)
    elif i[1]['bt'] == 'Sedan':
        data['seater'] = data['seater'].apply(lambda x: '5' if pd.isnull(x) else x)
    elif i[1]['bt'] == 'Minivans':
        data['seater'] = data['seater'].apply(lambda x: '6' if pd.isnull(x) else x)
    elif i[1]['bt'] == 'Pickup Trucks':
        data['seater'] = data['seater'].apply(lambda x: '6' if pd.isnull(x) else x)
    elif i[1]['bt'] == 'wagon':
        data['seater'] = data['seater'].apply(lambda x: '5' if pd.isnull(x) else x)
    elif i[1]['bt'] == 'Coupe':
        data['seater'] = data['seater'].apply(lambda x: '5' if pd.isnull(x) else x)
    else:
        break
```
